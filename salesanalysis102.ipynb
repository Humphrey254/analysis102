```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('sales.csv')
df.head()
```

### year with many sales and less sales

```python

df['total_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales'] + df['Global_Sales']
sales= pd.DataFrame(df.groupby('Year').sum()['total_sales'])
sales = sales.sort_values( 'total_sales', ascending=False)
sales[:]
```

#### top 10 product sold the most 

```python
best_product= pd.DataFrame(df.groupby('Platform').sum()['total_sales'])
best_product =best_product.sort_values('total_sales',ascending=False)
best_product[:10]
```

### which publisher was used the most 

```python
pub = pd.DataFrame(df.groupby('Publisher').sum()['total_sales'])
pub = pub.sort_values('total_sales',ascending=False)
pub[:10]
```
